Suppose your little sibling wants your help to check his math homework. He is done with his homework but wants you to see if all his results are correct. Since the student with all correct results gets 3 stars. However, you want your brother to check this on his own. So, you design a calculator for him in python. You could have given your scientific calculator but you wanted to give him a basic calculator and also wanted to see if you can even design one.
Subtasks:
Create a class called Calculator.
Your class shall have 1 constructor and 4 methods, namely add, subtract, multiply and divide.
Now, create an object of your class. After creating an object, it should print “Let’s Calculate!”
Then take 3 inputs from the user: first value, operator, second value
Now based on the given operator, call the required method and print the result.
Sample Input:
1
+
2
Sample Output:
Let’s Calculate!
Value 1: 1
Operator: +
Value 2: 2
Result:  3


```python
class Calculator:
    def __init__(self):
        print("Let's Calculate!")

    def add(self,num1,num2):
        return num1+num2
    
    def substract(self,num1,num2):
        return num1-num2
    
    def multiply(self,num1,num2):
        return num1*num2
    
    def devide(self,num1,num2):
        return num1/num2

calc1 = Calculator()

val1 = int(input())
print(f"Value 1: {val1}")

operator = input()
print(f"Operator: {operator}")

val2=int(input())
print(f"Value 2: {val2}")

if operator=="+":
    print(f"Result: {calc1.add(val1,val2)}")
elif operator=="-":
    print(f"Result: {calc1.substract(val1,val2)}")
elif operator=="*":
    print(f"Result: {calc1.multiply(val1,val2)}")
else:
    print(f"Result: {calc1.devide(val1,val2)}")
```

Implement the design of the Student class so that the following output is produced:
Assume the credit for each course to be 3. For example: [3.3,4] can be calculated as: 
CGPA = ((3.3 * 3) + (4 * 3)) / 6 
[Here, for each course, the grade point is multiplied by 3. Total credit is the number of courses multiplied by 3. Since the example has 2 courses, therefore a total of 6 credits]
CGPA = sum of individual (grade point * credit) / total credit
Academic Standing Rule: [CGPA>3.80 Highest Distinction, CGPA>3.65 High Distinction, CGPA>3.50 Distinction, CGPA>2.00 Satisfactory, CGPA<2.00 Can’t Graduate]

# Write your code here

```python
class Student:
    def __init__(self,name,st_id,course,cgpa_l):
        self.name=name
        self.st_id=st_id
        self.course=course
        self.cgpa_l=cgpa_l

    def calculate_CGPA(self):
        cgpa=0
        for i in self.cgpa_l:
            cgpa+=i*3

        self.total=cgpa/len(self.cgpa_l*3)

        if self.total>3.80:
            self.standing = "Your academic standing is 'Highest Distinction'"
        elif self.total>3.65:
            self.standing = "'Your academic standing is High Distinction'"
        elif self.total>3.50:
            self.standing = "Your academic standing is 'Distinction'"
        elif self.total>2.00:
            self.standing = "Your academic standing is 'Satisfactory'"
        else:
            self.standing = "Sorry you can't Graduat"
    def print_details(self):
        print(f"""Name: "{self.name}", ID: {self.st_id}
Department: {self.course} 
CGPA: {self.total}
{self.standing}""")

        


s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================") 
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
class Spotify:
    def __init__(self,songs):
        self.songs=songs
        self.total=len(songs)

    def add_to_playlist(self,song):
        self.songs.append(song)
        self.total=len(self.songs)
    
    def playing_number(self,num):
        if len(self.songs)<num:
            return(f"{num} number song not found. Your playlist has {self.total} songs only.")
        else:
#             return(f"""########################## #! better
# Playing {num} number song for you
# Song name: {self.songs[num-1]}""")
            return (f"##########################\nPlaying {num} song for you\nSong name:{self.songs[num-1]}")
                      

user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```
